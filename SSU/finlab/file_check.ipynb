```python
import pandas as pd
from IPython.display import display
```

```python
df1 = pd.read_csv('KOSPI_BETA_MONTHLY.csv',encoding = "ISO-8859-1")
df2 = pd.read_csv('KOSPI_SILH_SUIK_RT_WITH_DIVIDEND_MONTHLY.csv',encoding = "ISO-8859-1")
df1.shape , df2.shape
```

```python
df1.index = df1.iloc[:,0]
df1 = df1.iloc[12:,1:]
df1.columns =  ['stock'+str(idx) for idx,val in enumerate(df1.columns)]
df1.fillna(0,inplace=True)
display(df1.tail())

df2.index = df2.iloc[:,0]
df2 = df2.iloc[12:,1:]
df2.columns =  ['stock'+str(idx) for idx,val in enumerate(df2.columns)]
df2.fillna(0,inplace=True)
display(df2.tail())

print(df1.shape , df2.shape)
```
