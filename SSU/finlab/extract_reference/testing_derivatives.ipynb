```python
import time
from selenium import webdriver

def crawling_derivatives(page_num=2):
    url = 'http://www.kafo.or.kr/board/list.asp'
    driver = webdriver.Chrome()
    driver.get(url)

    year_ls = []
    paper_name = []

    for page in range(1,page_num):
        page_link = 'http://www.kafo.or.kr/board/list.asp?page={}&SearchKey=&SearchStr='.format(page)
        driver.get(page_link)
        driver.execute_script('window.scrollTo(200,500)')
        for i in range(1,10+1):
            driver.execute_script('window.scrollTo(200,500);')
            year_ls.append(driver.find_element_by_css_selector('#content_s > div > form > table > tbody > tr:nth-child({}) > td:nth-child(4)'.format(i)).text.split('.')[0])

            if year_ls[-1] > '2008':
                time.sleep(1)
                driver.find_element_by_css_selector('#content_s > div > form > table > tbody > tr:nth-child({}) > td:nth-child(2) > a'.format(i)).click()
                for j in range(1,10+1):
                    driver.execute_script('window.scrollTo(200,1000);')
                    try : 
    #                     paper_name.append(driver.find_element_by_css_selector('#content_s > div > form > table > tbody > tr:nth-child(4) > td > a:nth-child({})'.format(i)).text)
                        driver.find_element_by_css_selector('#content_s > div > form > table > tbody > tr:nth-child(4) > td > a:nth-child({})'.format(j)).click()
                    except : pass
                driver.get(page_link)

            else : break
    driver.close()
```

```python
crawling_derivatives(1000)
```
