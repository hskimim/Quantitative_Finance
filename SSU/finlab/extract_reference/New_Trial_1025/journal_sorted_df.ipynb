```python
import pandas as pd
import re
```

```python
df = pd.read_csv('new_tuned_references.csv')
df.tail()
```

# 저널별 논문 수 , 피인용 논문 수 파악하는 데이터 프레임 생성

```python
df['cited_count_infor'] = [int(','.join(re.findall('\d',val)).replace(",",'')) for val in df['cited_infor']]
```

```python
df.tail(1)
```

```python
first_groupby_df = df.groupby(by='institution_infor').aggregate({'paper_name_infor':'unique','char_refer':'count'})
first_groupby_df
```

```python
first_groupby_df['citing_paper_count'] = [len(i) for i in first_groupby_df['paper_name_infor']]
```

```python
groupby_df = first_groupby_df.iloc[:,1:]
groupby_df.columns = ['cited_paper_num','citing_paper_num']
groupby_df['cited_paper_per_citing_paper'] = groupby_df['cited_paper_num'] / groupby_df['citing_paper_num']
groupby_df
```

# 연도별 인용 , 피인용 논문 갯수 파악하는 데이터 프레임

```python
df.tail(1)
```

```python
first_groupby_df = df.groupby(by='year_infor').aggregate({'paper_name_infor':'unique','char_refer':'count'})
first_groupby_df
```

```python
first_groupby_df['citing_paper_count'] = [len(i) for i in first_groupby_df['paper_name_infor']]
```

```python
groupby_df = first_groupby_df.iloc[:,1:]
groupby_df.columns = ['cited_paper_num','citing_paper_num']
groupby_df['cited_paper_per_citing_paper'] = groupby_df['cited_paper_num'] / groupby_df['citing_paper_num']
groupby_df
```

# 각 저널별로 몇 년도 부터 몇 년도까지의 인용논문이 있는지 파악하는 데이터 프레임

```python
df.tail(1)
```

```python
year_infor_df = df.groupby('institution_infor').aggregate({'year_infor':'unique'})
year_infor_df['year_infor_tuned'] = [set(i) for i in year_infor_df['year_infor']]
year_infor_df['count_year_infor'] = [len(i) for i in year_infor_df['year_infor_tuned']]
year_infor_df = year_infor_df.iloc[:,1:]
year_infor_df
```
