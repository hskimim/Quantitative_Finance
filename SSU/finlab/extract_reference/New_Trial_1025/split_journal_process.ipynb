```python
import pandas as pd
from IPython.display import display
import numpy as np
import re
from collections import Counter
import nltk
```

```python
df = pd.read_csv('tuned_dataframe.csv')
```

```python
journal_name_ls = list(pd.read_csv('journal_name_ls.csv')['0'].values)
```

```python
for idx,val in enumerate(journal_name_ls) : 
    if val[-1] == ' ':
        journal_name_ls[idx] = val[:-1]
```

```python
journal_name_ls = list(set(journal_name_ls))
```

```python
journal_name_ls
```

```python
journal_name_ls = [i for i in journal_name_ls if len(i) >2]
```

```python
split_ls = []
for idx,val in enumerate(df['paper_refer']):
    for j in journal_name_ls:
        if j in val:
            split_ls.append((idx,j))
            print(j)
            break
            
```

```python
split_ls
```

```python
%%time
for idx,val in enumerate(df['paper_refer']) : 
    if idx in [i[0] for i in split_ls] :
        loc = [i[0] for i in split_ls].index(idx)
        df['journal_refer'][idx] = val[re.search(split_ls[loc][1],df['paper_refer'][idx]).start() : ]
        df['paper_refer'][idx] = val[ : re.search(split_ls[loc][1],df['paper_refer'][idx]).start() ]
```
