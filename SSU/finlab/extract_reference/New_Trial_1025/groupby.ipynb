```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('new_tuned_references.csv')
df.tail()
```

```python
df['year_infor'] = [str(i).split('.')[0] for i in df['year_infor']]
df['year_refer'] = [str(i).split('.')[0] for i in df['year_refer']]
```

```python
groupby_df1 = df.groupby(['journal_infor','year_infor']).aggregate({'paper_name_infor':'unique','paper_refer':'count'})
```

```python
groupby_df2 = df.groupby(['institution_infor','year_infor']).aggregate({'paper_name_infor':'unique','paper_refer':'count'})
```

```python
groupby_df1['paper_name_infor'] = [len(i) for i in groupby_df1['paper_name_infor']]
groupby_df2['paper_name_infor'] = [len(i) for i in groupby_df2['paper_name_infor']]
```

```python
groupby_df1['ratio'] = groupby_df1['paper_refer'] / groupby_df1['paper_name_infor']
groupby_df2['ratio'] = groupby_df2['paper_refer'] / groupby_df2['paper_name_infor']
```
