```python
import pandas as pd
import re
import numpy as np
```

```python
df = pd.read_csv('new_tuned_references.csv')
```

```python
df.tail()
```

```python
df['year_infor'] = [str(i).split('.')[0] for i in df['year_infor']]
df['year_refer'] = [str(i).split('.')[0] for i in df['year_refer']]
```

```python
set([re.search('pp.',val).end() for val in df['page_infor'].values])
```

```python
df['page_count_infor'] = [int(i[4:].split('~')[1]) - int(i[4:].split('~')[0]) for i in df['page_infor']]
```

```python
df.to_csv("new_tuned_references.csv",index=False)
```

```python
df[df['paper_name_infor'] == '1980～2004년 동안의 증시부양정책 및 증시규제정책의 실효성']
```

```python
refer_per_paper = df.groupby(['paper_name_infor']).aggregate({'institution_infor':'count','page_count_infor':'mean'})
refer_per_paper.columns = ['reference_num','paper_page_num']
refer_per_paper['ratio'] = np.round(refer_per_paper['paper_page_num'] / refer_per_paper['reference_num'],2)
refer_per_paper
```
