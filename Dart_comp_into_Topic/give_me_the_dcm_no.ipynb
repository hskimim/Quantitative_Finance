```python
from personal_function import *
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
def find_between( s, first, last ):
    start = s.rfind( first ) + len( first )
    end = s.index( last, start )
    return s[start:end]

# rfind() returns the last index where the substring str is found, 
# or -1 if no such index exists, optionally restricting the search to string[beg:end].
```

```python
industrial_url = pd.read_csv('industrial_url_kospi.csv')
```

```python
recently_issued_df = pd.DataFrame(columns=['industrial_url','company_name','when_is_issued'])
for name in list(set(industrial_url['company_name'])):
    recently_issued_df = recently_issued_df.append(industrial_url[industrial_url['company_name']==name].iloc[0])
```

```python
ls = []
industrial_source_url = []

for i in range(len(recently_issued_df)):
    url = recently_issued_df['industrial_url'][i]
    r = requests.get(url)
    dom = BeautifulSoup(r.text,'lxml')
    
    try : 
        start_str = 'text: "I. 회사의 개요",'
        end_str = "}"
        doc = find_between(r.text , start_str , end_str)

        second_start_str = '{viewDoc('
        end_str = ", 'dart3.xsd');"

        doc2 = find_between(doc , second_start_str , end_str)
        doc2 = doc2.replace("'",'').replace(' ','').split(',')
        source_url = 'http://dart.fss.or.kr/report/viewer.do?rcpNo={}&dcmNo={}&eleId={}&offset={}&length={}&dtd=dart3.xsd'.format(doc2[0],doc2[1],doc2[2],doc2[3],doc2[4])
        industrial_source_url.append(source_url)
    except : ls.append(i)
```
